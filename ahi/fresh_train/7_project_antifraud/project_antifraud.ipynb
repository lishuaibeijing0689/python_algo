```python
# 把类别字段编码为数值
cols_encode = [
    'adunitshowid',
 'apptype',
 'carrier',
 'city',
 'dvctype',
 'h',
 'idfamd5',
 'ip_carrier',
 'ip_city',
 'ip_prefix_20',
 'ip_prefix_22',
 'ip_prefix_24',
 'lan',
 'make',
 'mediashowid',
 'model',
 'ntt',
 'openudidmd5',
 'orientation',
 'os',
 'osv',
 'pkgname',
 'ppi',
 'province',
 'reqrealip_carrier',
 'reqrealip_city',
 'reqrealip_prefix_20',
 'reqrealip_prefix_22',
 'reqrealip_prefix_24',
 'ver',
 'w'
]

df_feature_param = df
for col in cols_encode:
    df_encode_tmp = spark.createDataFrame(df.select(col).drop_duplicates().rdd.map(lambda x: x[0]).zipWithIndex(), schema=[col, col+"_encode"])
    df_feature_param = df_feature_param.join(df_encode_tmp, on=col, how="left").drop(col)
    print col
```

```python
df_global_cnt = df.select(cols_global_cnt + ["sid", "timestamp"])

for col in cols_global_cnt:
    df_tmp = df.select(col).groupby(col).count().withColumnRenamed("count", col+"_global_cnt")
    df_global_cnt = df_global_cnt.join(df_tmp, on=col, how="left").drop(col)
    print col
```

### 生成 训练集、验证集和测试集特征矩阵

```python
df_train_sid_label = spark.read.parquet("/data/fresh_train/project_antifraud/train_set_label")
df_train_sid_label.show(5)
```

```python
df_train_sid_label.filter(functions.col("label")==1).count()
```

```python
df_test_sid_label = spark.read.parquet("/data/fresh_train/project_antifraud/test_set_label")
df_test_sid_label.show(5)
```

```python
df_test_sid_label.filter(functions.col("label")==1).count()
```

```python
ts_cut = 1559813760751 # 训练集时间戳小于ts_cut作为train，大于ts_cut的作为valid   todo 这个数值可以自定义
```

```python
df_feature_train = df_train_sid_label.join(df_feature_param, on="sid", how="left").filter(functions.col("timestamp")<ts_cut).drop("timestamp")
df_feature_valid = df_train_sid_label.join(df_feature_param, on="sid", how="left").filter(functions.col("timestamp")>=ts_cut).drop("timestamp")
df_feature_test = df_test_sid_label.join(df_feature_param, on="sid", how="left").drop("timestamp")
```

```python
pdf_feature_train = df_feature_train.toPandas()
pdf_feature_valid = df_feature_valid.toPandas()
pdf_feature_test = df_feature_test.toPandas()

pdf_feature_train.shape, pdf_feature_valid.shape, pdf_feature_test.shape
```

```python
df_feature_global_cnt_train = df_train_sid_label.drop("label").join(df_global_cnt, on="sid", how="left").filter(functions.col("timestamp")<ts_cut).drop("timestamp")
df_feature_global_cnt_valid = df_train_sid_label.drop("label").join(df_global_cnt, on="sid", how="left").filter(functions.col("timestamp")>=ts_cut).drop("timestamp")
df_feature_global_cnt_test = df_test_sid_label.drop("label").join(df_global_cnt, on="sid", how="left").drop("timestamp")
```

```python
pdf_feature_global_cnt_train = df_feature_global_cnt_train.toPandas()
pdf_feature_global_cnt_valid = df_feature_global_cnt_valid.toPandas()
pdf_feature_global_cnt_test = df_feature_global_cnt_test.toPandas()

pdf_feature_global_cnt_train.shape, pdf_feature_global_cnt_valid.shape, pdf_feature_global_cnt_test.shape
```

```python
# 把两个特征矩阵join起来，生成一个完整的特征矩阵
pdf_train = pd.merge(pdf_feature_train, pdf_feature_global_cnt_train, on="sid", how="inner")
pdf_valid = pd.merge(pdf_feature_valid, pdf_feature_global_cnt_valid, on="sid", how="inner")
pdf_test = pd.merge(pdf_feature_test, pdf_feature_global_cnt_test, on="sid", how="inner")

pdf_train.shape, pdf_valid.shape, pdf_test.shape
```

```python
feature_names = list(set(pdf_train.columns) - set(["label", "sid"]))
print len(feature_names)

x_train = pdf_train[feature_names]
y_train = pdf_train["label"].astype("int")

x_valid = pdf_valid[feature_names]
y_valid = pdf_valid["label"].astype("int")

x_test = pdf_test[feature_names]
y_test = pdf_test["label"].astype("int")

x_train.shape, y_train.shape, x_valid.shape, y_valid.shape, x_test.shape, y_test.shape
```

```python
cat_features = [x for x in pdf_train.columns if "encode" in x]
len(cat_features)
```

```python
params = {
    'boosting_type': 'gbdt',
    'objective': 'binary',
    'metric': 'auc',
    'num_leaves': 31,
    'learning_rate': 0.1,
    'feature_fraction': 0.7,
    'bagging_fraction': 0.7,
    'bagging_freq': 2,
    'lambda_l1': 0.25,
    'lambda_l2': 0.25,
    'verbose': 0,
    'max_depth': 6,
    'n_estimators': 5000,
#     'min_child_weight': 5,
    'random_state': 2019,
    'category_feature': cat_features,  # 指定类别特征
    'n_jobs': 32
}
```

```python
lgb_model = lgb.LGBMClassifier(**params)

lgb_model.fit(x_train, 
              y_train, 
              eval_set = [(x_train, y_train), (x_valid, y_valid)], 
              early_stopping_rounds=50, 
              feature_name = feature_names)
```

```python
lgb.plot_importance(lgb_model, figsize=(18, 20), max_num_features=20)
```

```python
params["n_estimators"] = lgb_model.best_iteration_ + 200
```
