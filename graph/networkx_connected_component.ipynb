```python
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
pointList = ['A','B','C','D','E','F','G']
linkList = [('A','B'),('B','C'),('C','D'),('E','F'),('F','G'),]
```

```python
G = nx.Graph()
# 转化为图结构
for node in pointList:
    G.add_node(node)

for link in linkList:
    G.add_edge(link[0], link[1])
```

```python
G.nodes
```

```python
G.edges
```

```python
nx.draw_networkx(G, with_labels=True)
```

```python
subplot = [121,122]
color =['y','g']
```

```python
for c in nx.connected_components(G): # 注意：nx.connected_components(G)是一个生成器，只能迭代一次
#     print(c)
   # 得到不连通的子集
    nodeSet = G.subgraph(c).nodes()
#     print(nodeSet)
   # 绘制子图
    subgraph = G.subgraph(c)
    plt.subplot(subplot[0])  # 第二整行
    nx.draw_networkx(subgraph, with_labels=True,node_color=color[0])
    color.pop(0)
    subplot.pop(0)
```
