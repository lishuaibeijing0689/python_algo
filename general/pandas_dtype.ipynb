```python
"""
在读取数据时指定dtype
"""
```

```python
import pandas as pd
```

```python
import sys
```

```python
dtype_dict = {
    "A": "float32",
    "B": "float32",
    "C": "float32",
    "D": "str",
    "E": "str",
    "F": "str"
}
```

```python
pdf = pd.read_csv("demo_csv.csv", header=0, dtype=dtype_dict)
```

```python
pdf
```

```python
pdf.dtypes
```

```python
sys.getsizeof(pdf)
```
