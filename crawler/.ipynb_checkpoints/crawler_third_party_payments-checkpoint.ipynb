```python
"""
下载人民银行第三方支付机构清单

url：
http://www.pbc.gov.cn/zhengwugongkai/127924/128041/2951606/1923625/1923629/d6d180ae/index1.html
http://www.pbc.gov.cn/zhengwugongkai/127924/128041/2951606/1923625/1923629/d6d180ae/index2.html
http://www.pbc.gov.cn/zhengwugongkai/127924/128041/2951606/1923625/1923629/d6d180ae/index3.html
...

对方网站安装了反爬虫，爬不下来
"""
```

```python
import urllib.request
```

```python
 def getHtml(url):
    html = urllib.request.urlopen(url).read()
    return html
 
def saveHtml(file_name, file_content):
    with open(file_name.replace('/', '_'), "wb") as f:
        #   写文件用bytes而不是str，所以要转码
        f.write(file_content)
```

```python
file_list = []
for i in range(1,13):
    f = "http://www.pbc.gov.cn/zhengwugongkai/127924/128041/2951606/1923625/1923629/d6d180ae/index%d.html" % i
    html = getHtml(f)
    saveHtml(f, html)
    file_list.append(f)
```

```python
import re
```

```python
# 示例
pattern = re.compile(r'\d+')   # 查找数字
result1 = pattern.findall('runoob 123 google 456')
result2 = pattern.findall('run88oob123google456', 0, 10)
 
print(result1)
print(result2)
```

```python
def get_all_payments():
    payments = []
    for i in range(12):
        with open("%d.html" % (i+1)) as f:
            html = f.read()
        pattern = re.compile(r"title=\".*\"")
        result = pattern.findall(html)
        payments += [i[7:-1] for i in result]
    return payments
```

```python
payments = get_all_payments()
len(payments), payments
```

```python
import pandas as pd
```

```python
pd.DataFrame(payments, columns=["payments"]).to_csv("third_party_payments.csv", index=False, header=False)
```
